#### BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.head.prettify())
```

```python
table_data = soup.find(id='ctl00_Contenido_tblAcciones')
```

```python
headers = []
for i in table_data.find_all('th'):
    title = i.text
    headers.append(title)
```

```python
print(headers)
```

```python
stocks_df = pd.DataFrame(columns = headers)
```

```python
for j in table_data.find_all('tr')[1:]:
    row_data = j.find_all('td')
    row = [tr.text for tr in row_data]
    length = len(stocks_df)
    stocks_df.loc[length] = row
```

```python
stocks_df
```

```python
stocks_df.to_csv('ValoresBolsaMadrid.csv')
```

```python
from selenium import webdriver
from selenium.webdriver import Safari
from selenium.webdriver.safari.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
import pandas as pd
```

```python
driver = webdriver.Safari()
```

```python
driver.get('https://www.bolsamadrid.es')
```

```python
driver.title
```

```python
driver.current_url
```

```python
acciones = driver.find_element_by_xpath('//div[@class="MenuLink MenuConOp Off"]/a[text()="Acciones"]')
acciones.click()
```

```python
driver.current_url
```

```python
sector = driver.find_element_by_xpath('//select[@id="SelSector"]/option[text()="Petróleo y Energía"]')
sector.click()
```

```python
consultar = driver.find_element_by_xpath('//input[@id="ctl00_Contenido_Consultar"]')
consultar.click()
```

```python
table = driver.find_element_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody')
```

```python
headings = []
for i in table.find_elements_by_xpath('.//tr'):
    for j in i.find_elements_by_xpath('./th'):
        headings.append(j.get_attribute('textContent'))

print(headings)       
```

```python
stocks_df2 = pd.DataFrame(columns = headings)
stocks_df2
```

```python
table_data = []
c = 0

for i in table.find_elements_by_xpath('.//tr'):
    for j in i.find_elements_by_xpath('./td'):
        table_data.append(j.get_attribute('textContent').strip())
    if table_data == []:
        continue
    stocks_df2.loc[c] = table_data
    c += 1
    table_data = []
```

```python
stocks_df2
```

```python
driver.quit()
```

```python
stocks_df2.to_csv('Petro_Energ.csv')
```
